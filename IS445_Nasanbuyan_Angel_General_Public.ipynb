## 1. Time Series Plot of pH Levels Over Time

This visualization will visualize how pH levels in the water change over time, providing insights into the water's acidity or alkalinity trends which are crucial for ecosystem health.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('BKB_WaterQualityData_2020084.csv')
df['Read_Date'] = pd.to_datetime(df['Read_Date'])  

plt.figure(figsize=(10, 6))
plt.plot(df['Read_Date'], df['pH (standard units)'], marker='o', linestyle='-', color='blue')
plt.title('pH Levels Over Time')
plt.xlabel('Date')
plt.ylabel('pH (standard units)')
plt.grid(True)
plt.show()
```

This visualization displays the changes in water pH levels over several years. This is also measured from various sites. pH is a crucial water quality parameter that affects aquatic life and indicates the water's acidic or alkaline condition. The plot helps in identifying trends, seasonal variations, or any anomalies in pH levels. The key observations I found are the pH level appears to fluctuate within the normal range and it looks like a stable water quality conditions. This also shows how it is important to monitor the pH level to keep checking the water quality. This graph provides a valuable insights into the water quality dynamics and assists in making informed decisions regarding water management and conservation efforts.


## 2. Bar Chart of Average Water Temperature by Year

This chart will help visualize the average water temperature each year.

```python

avg_temp_by_year = df.groupby('Year')['Water Temp (?C)'].mean()

avg_temp_by_year.plot(kind='bar', color='teal', figsize=(10, 6))
plt.title('Average Water Temperature by Year')
plt.xlabel('Year')
plt.ylabel('Average Water Temperature (°C)')
plt.xticks(rotation=45)
plt.show()
```

This bar chart represents the trends in average water temperatures recorded annually. Monitoring water temperature is crucial as it influences the ecological balance of aquatic environments. This visualization highlights the potential areas of concern and may help in prioritizing efforts towards maintaining or improving the health of aquatic environments.


## 3. Scatter Plot of Salinity vs. Dissolved Oxygen 

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Salinity (ppt)'], df['Dissolved Oxygen (mg/L)'], color='magenta', alpha=0.5)
plt.title('Salinity vs. Dissolved Oxygen Levels')
plt.xlabel('Salinity (ppt)')
plt.ylabel('Dissolved Oxygen (mg/L)')
plt.grid(True)
plt.show()
```

This scatter plot examines the relationship between salinity (measured in parts per thousand, ppt) and dissolved oxygen levels (measured in mg/L) in water samples. I wanted to understand this relationship as it is crucial as it impacts aquatic life. Different species require different oxygen levels, and because of that salinity can significantly influence oxygen solubility.
My initial thought was that higher salinity might correlate with lower dissolved oxygen levels, since saltwater typically holds less oxygen than freshwater. And there appears that there is no clear correlation as my initial thought. This suggest that other factors can influence dissolved oxygen.
